### CHAPTER 08 - GROUPING VS PIVOT TABLE VS CROSSTAB
- grouping the data according to the categories and apply a function to the categories. 
- Pivot table in pandas is an excellent tool to summarize one or more numeric variable based on two other categorical variables
- Cross tabulation is used to quantitatively analyze the relationship between multiple variables.

```python
insurance_details=pd.read_csv(r"C:\Users\PRIYANKA\Data_Science_14_april_jhon\data_Sets\insurance.csv")
insurance_details
```

## CLUE TO USE GROUPBY FUNCTION

DISCRETE VS CONTINUES DATA TYPES

**DISCRETE** - COUNTABLE BUT NOT MEASURABLE(NO UNITS) AND CANNOT BE SEGREGATTED


**CONTINUOUS** - MEASURABLE(UNITS), CAN BE SEGREGATTED

groupby(discrete data)[cont data]

#### 8.1 ON AN AVERAGE WHICH REGION IS GETTING MORE INSURANCE CHARGES


```python
insurance_details.groupby(by='region')['charges'].mean().sort_values(ascending=False)
```

```python
insurance_details.groupby(by='region')['charges'].mean().sort_values(ascending=True)
```

#### 8.2  ON AN AVERAGE WHICH REGION IS GETTING MAX INSURANCE CHARGES

```python
insurance_details.groupby(by='region')['charges'].max().sort_values(ascending=False)
```

```python
####  ON AN AVERAGE WHICH GENDER IS GETTING MAX INSURANCE CHARGES
insurance_details.groupby(by='sex')['charges'].mean().sort_values(ascending=False).round()
```

#### 8.3 BASED ON REGION AND SMOKERS WHAT IS THE AVERAGE INSURANCE CHARGES?

```python
insurance_details.groupby(by=['smoker','region'])['charges'].mean().sort_values(ascending=False).round()
```

```python
insurance_details.groupby(by=['smoker','region','children'])['charges'].mean().sort_values(ascending=False).round()
```

```python
pd.pivot_table(data = insurance_details,values = 'charges',index='region',columns=['smoker','children'],aggfunc='mean').round()
```

```python
pd.pivot_table(data = insurance_details,values = 'charges',index='region',columns=['smoker'],aggfunc='mean').round().sort_values(by='yes',ascending=False)
```
